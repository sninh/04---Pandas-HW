```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Find total amount pf players
total_players = len(purchase_data["SN"].value_counts())
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Set values for unique items, average price, number of purchases, and total revenue
unique_items = len(purchase_data["Item ID"].value_counts())
average_price = purchase_data["Price"].mean()

number_purchase = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

#put these values into a data frame
analysis_df = pd.DataFrame({"Number of Unique Items":[unique_items],
                          "Average Price":[average_price],
                           "Number of Purchases":[number_purchase],
                           "Total Revenue":[total_revenue]})
analysis_df["Average Price"] = analysis_df["Average Price"].map("${:.2f}".format)
analysis_df["Total Revenue"] = analysis_df["Total Revenue"].map("${:,.2f}".format)
analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#group the values by gender
genders = purchase_data.groupby("Gender")
#find the count of unique users in this group
unique_users = genders.nunique()["SN"]
#calc the % of users based off gender
gender_percentage = (unique_users / total_players)*100
#put these values into a data frame
gender_demographic_df = pd.DataFrame({"Total Count":unique_users, 
                                      "Percentage of Players":gender_percentage})
#Format monetary values into $
gender_demographic_df["Percentage of Players"] = gender_demographic_df["Percentage of Players"].map("{:.2f}%".format)
gender_demographic_df.index.name = None
gender_demographic_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Set values for average person, average price, number of purchases, and total revenue from above group
average_cost = genders.mean()["Price"]
purchase_count = genders["SN"].count()
total_purchase = (genders["Price"].sum())
average_person = total_purchase / unique_users
#Put this data into a DF
purchasing_analysis_df = pd.DataFrame({"Purchase Count":purchase_count,
                                      "Average Purchase Price":average_cost,
                                      "Total Purchase Value":total_purchase,
                                      "Avg Total Purchase per Person": average_person})
#Format monetary values into $
purchasing_analysis_df['Average Purchase Price'] = purchasing_analysis_df['Average Purchase Price'].map("${:.2f}".format)
purchasing_analysis_df['Total Purchase Value'] = purchasing_analysis_df['Total Purchase Value'].map("${:,.2f}".format)
purchasing_analysis_df['Avg Total Purchase per Person'] = purchasing_analysis_df['Avg Total Purchase per Person'].map("${:.2f}".format)
purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create age bins
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_label = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#Group data into DF
sn_df = purchase_data
age_df = sn_df
age_df["Age group"] = pd.cut(age_df["Age"], age_bins, labels=age_label)
age_group_df=age_df.groupby("Age group")
#calc total counts and % of players
total_age = age_group_df["SN"].nunique()
percentage_age = total_age / total_players * 100
#add into DF
age_answers = pd.DataFrame({"Total Count":total_age,
                           "Percentage of Players":percentage_age})
#Format percentages into %
age_answers["Percentage of Players"] = age_answers["Percentage of Players"].map("{:.2f}%".format)
age_answers
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#make new grouping by SN
spenders = purchase_data.groupby("SN")
#set variables
user_purchases = spenders["Price"].count()
user_average = spenders["Price"].mean()
user_total = spenders["Price"].sum()
#put variables in DF
spenders_df = pd.DataFrame({"Purchase Count":user_purchases, 
                "Average Purchase Price":user_average,
               "Total Purchase Value":user_total})
spenders_df=spenders_df.sort_values(by=["Total Purchase Value"], ascending=False)
#formatting
spenders_df["Average Purchase Price"] = spenders_df['Average Purchase Price'].map("${:,.2f}".format)
spenders_df["Total Purchase Value"] = spenders_df["Total Purchase Value"].map("${:,.2f}".format)
spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#make new grouping by items
popular = purchase_data[["Item ID", "Item Name", "Price"]]
popular_groups = popular.groupby(["Item ID", "Item Name"])
#set variables
pop_purchase_count = popular_groups["Item ID"].count()
pop_prices = popular_groups["Price"].mean()
pop_total_value = popular_groups["Price"].sum()
#put variables in DF
popular_df = pd.DataFrame({"Purchase Count":pop_purchase_count,
                          "Item Price":pop_prices,
                          "Total Purchase Value":pop_total_value})
sorted_popular_df=popular_df.sort_values(by=["Purchase Count"], ascending=False)
#formatting
sorted_popular_df["Item Price"] = popular_df["Item Price"].map("${:,.2f}".format)
sorted_popular_df["Total Purchase Value"] = popular_df["Total Purchase Value"].map("${:,.2f}".format)
sorted_popular_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting

```python
#sort above table
profitable_df = popular_df.sort_values(by=["Total Purchase Value"], ascending=False)
#format
profitable_df["Item Price"] = profitable_df["Item Price"].map("${:,.2f}".format)
profitable_df["Total Purchase Value"] = profitable_df["Total Purchase Value"].map("${:,.2f}".format)
profitable_df.head()
```
